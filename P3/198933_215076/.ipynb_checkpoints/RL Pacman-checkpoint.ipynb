Para facilitar a demontração dos resultados neste notebook, uitlizaremos a seguinte função:

```python
import numpy as np
from pacman import *
import matplotlib.pyplot as plt

def plot_results(numTraining=50,numGames=10,board='smallClassic',
                 alpha=0.15,gamma=0.1,epsilon=0.5,agent='ApproximateQPolicy'):
    """ número de passos máximo é definido em __init__ de Game """
    
    if __name__ == '__main__':
        
        numGames+=numTraining
    
        agentargs=[]
        agentargs.append('alpha={},gamma={},epsilon={},numTraining={},numGames={}'.format(
                        alpha,gamma,epsilon,numTraining,numGames))
        
        for i in agentargs:
            args = readCommand(['-a',i,'-l',board,'-p',agent,'-z',3,'-q',
                                '-x',str(numTraining),'-n',str(numGames)])
            game = runGames(**args)

            scores = [k.state.getScore() for k in game]
            plt.plot(np.arange(len(scores)) + 1, scores)    
            plt.title(str(i))
            plt.show()
```

## Funtion Aproximation based on Q-Learning

  Como explicado anteriormente, decidi implementar um aproximador de função linear para resover o problema. **Todas as escolhas de modelagem do problema mencionadas anteriormente foram mantidas**. A única diferença foi que, ao invés de redefinir o estado no Q-learning, utilizei um extrator de features. 

##### extração de features

  Entre as features escolhidas, me baseei nas features mencionadas pelo artigo *"Reinforcement Learning in Pacman"*, disponibilizado pela professora, o qual relata quais foram as features mais importantes nos experimentos do grupo de pesquisa. Essas *features* são: número de fantasmas assustados e ativos a um e a dois passos de distância, distância a comida mais próxima, e se o Pacman comeu a comida se não tiver fantasmas por perto.
  
  Além dessas *features*, adicionei algumas que poderiam aumentar as chances do agente procurar comer mais fantasmas, e consequentemente aumentar a pontuação. Adicionei as features que verificam se o agente comeu uma cápsula, uma *feature* que relaciona a distância do fantasma ao pacman com seu tempo restante como fantasma, e uma feature para avaliar se o pacman está seguindo algum fantasma. 
   
  Para o smallClassic, mediumClassic e originalClassic, foi possível que o modelo convergisse com poucos treinamentos utilizando este aproximador de função e hiperparâmetros pré-definidos. Abaixo, seguem as features implementadas para este aproximador de função. 

```python
class Extractor:
    def getFeatures(self,state,action):
        food = state.getFood()
        walls = state.getWalls()
        ghosts = state.getGhostPositions()

        features = util.Counter()
        
        # initialize bias
        features["bias"] = 1.0

        # compute the location of pacman after he takes the action
        x, y = state.getPacmanPosition()
        dx, dy = Actions.directionToVector(action)
        next_x, next_y = int(x + dx), int(y + dy)
        
        # compute legal forward actions
        next_actions=[Actions.getLegalNeighbors((next_x,next_y), walls)]
        
        # finds if there is closed scared ghosts 1 and 2 steps away: 
        features["#-of-scared-ghosts-1-step-away"] = sum((next_x, next_y) in 
                                                         Actions.getLegalNeighbors(ghosts[g], walls) 
                                                         for g in range(len(ghosts)) 
                                                         if ghostScaredTime(g+1,state) >0)
        features["#-of-scared-ghosts-2-steps-away"] = sum(a in Actions.getLegalNeighbors(ghosts[g], walls) 
                                                          for a in next_actions for g in range(len(ghosts)) 
                                                          if ghostScaredTime(g+1,state) >0)

        # finds if there is activ ghost nearby:
        features["#-of-active-ghosts-1-step-away"] = sum((next_x, next_y) in 
                                                         Actions.getLegalNeighbors(ghosts[g], walls) 
                                                         for g in range(len(ghosts)) 
                                                         if ghostScaredTime(g+1,state) <= 0)
        features["#-of-active-ghosts-2-steps-away"] = sum(a in Actions.getLegalNeighbors(ghosts[g], walls) 
                                                          for a in next_actions for g in range(len(ghosts)) 
                                                          if ghostScaredTime(g+1,state) <=0)

        # if there is no danger nearby, reward if pacman eats capsule:
        if not features["#-of-active-ghosts-1-step-away"] and bool((next_x,next_y) in state.getCapsules()):
            features['has-capsule-1-step-away']=1.0
        
        # if there is no danger nearby, reward if pacman eats fruit:
        if not features["#-of-active-ghosts-1-step-away"] and food[next_x][next_y]:
            features["eats-food"] = 1.0
        
        # find closest food:
        fruit = Graph.getClosestPos((next_x, next_y), walls, food)
        if fruit is not None:
            dist=fruit.dist
            features["closest-food"] = float(dist) / (walls.width * walls.height)
        
        # find closest scared ghost:
        scared_ghosts=[ghosts[s] for s in range(len(ghosts)) if ghostScaredTime(s+1,state)>0]
        scared_ghost=Graph.getClosestPos((next_x, next_y), walls, scared_ghosts)
        if scared_ghost != None:
            scared_ghost_dist=scared_ghost.dist
            scared_ghost_index=scared_ghost.ghost_id 
            scared_ghost_dir=scared_ghost.dir
            features["scaredTime-ghostDist"] = (float(ghostScaredTime(scared_ghost_index+1,state))*0.5 - 
                                               float(scared_ghost_dist) / (walls.width * walls.height))
            features["run-to-catch-scared-ghost"]=float(int(action==scared_ghost_dir)/(
                                                  scared_ghost_dist+1))*0.5
            
        features.divideAll(10.0)
        return features
```

## Avaliação de Hiperparâmetros

  Como visto, foi possível fazer com que o agente aprendesse com nosso extrator de features. No entanto, para finalizar este trabalho, serão demonstrados curtos experimentos avaliando a performance de outros hiperparâmetros. Devido ao tempo dos experimentos, especialmente no caso do tabuleiro 'originalClassic', farei treinamentos curtos. 

  O objetivo é avaliar se com outros hiperparâmetros é possível obter um agente que aprenda a obter **scores maiores**. Não será avaliado diretamente se agente aprendeu a vencer o jogo, por ter pouco treinamento. Será considerado primordiamente a **média**, a **mediana** e o **desvio padrão** dos resultados.
  
Desativarei a opção de ler e salvar os pesos das features, e plotar os gráficos, para acelerar o processo e deixá-lo mais limpo e visível. 



Com a função abaixo, será mais fácil de realizar os testes.


```python
import numpy as np
from pacman import *
import matplotlib.pyplot as plt

def isBestVal(median,mean,std,curr_val):
    comp=(median+mean-std)/2
    if comp>curr_val:
        curr_val=comp
        return comp
    return False

def print_results(numTraining=50,numGames=10,board='smallClassic',
                 alpha=[0.15,0.2],gamma=[0.1,0.2],epsilon=[0.5,0.2],agent='ApproximateQPolicy'):
    """ número de passos máximo é definido em __init__ de Game """
    
    if __name__ == '__main__':
        
        numGames+=numTraining
    
        agentargs=[]
        for i in (alpha):
            for j in (gamma):
                for k in (epsilon):
                    i=float(i)
                    j=float(j)
                    k=float(k)
                    agentargs.append('alpha={},gamma={},epsilon={},numTraining={},numGames={}'.format(
                    i,j,k,numTraining,numGames))
        max_args=None
        max_val=-1000
        for arg in agentargs:
            print("******************************************************************************")
            print(str(arg))
            args = readCommand(['-a',arg,'-l',board,'-p',agent,'-q',
                                '-x',str(numTraining),'-n',str(numGames)])
            game = runGames(**args)
            scores = [k.state.getScore() for k in game]
            
            median=np.median(scores)
            mean=np.mean(scores)
            std=np.std(scores)
            print("median of scores: {}\nstandard deviation of scores: {}".format(
                  np.median(scores), np.std(scores)))
            best_val=isBestVal(median,mean,std,max_val)
            if best_val:
                max_val=best_val
                max_args=arg
            
        print("best hyperparameters: {}".format(max_args))
```

##### smallClassic

Começaremos com um teste simples de verificação dos hiperparâmetros. Começaremos procurando por hiperparâmetros que tragam melhor resultado para os testes de smallClassic:

```python
alpha=np.arange(0.1,1,0.27)
gamma=np.arange(0.1,1,0.27)
eps  =np.arange(0.1,1,0.27)

print_results(5,5,'smallClassic',alpha,gamma,eps)
```

  Com número de treinamentos suficientes, priorizaríamos altas médias e medianas (mais que 970), e baixo desvio padrão (menos que 50). 
  No entanto, como temos pouco treinamento para cada conjunto de parâmetros, optou-se por salvar o resultado do conjunto de hiperparâmetros com maior valor de função *(média + mediana - desvio padrão)/2*.
  
Dados os respectivos valores, encontramos os hiperparâmetros dentro de nosso critério, dado por: alpha=0.64,gamma=0.64,epsilon=0.37.
Possui mediana 982.0, média 972.4 e desvio padrão 14.86.
